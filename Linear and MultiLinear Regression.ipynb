# Model 1

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

# Loading dataset and checking info

```python
df=pd.read_csv("houseprice.csv")
df.head()
```

```python
X=df.iloc[:,1].values.reshape(-1,1)
y=df.iloc[:,-1].values.reshape(-1,1)
```

# Pair Plot

```python
plt.scatter(X,y,color='r')
plt.xlabel('LotArea')
plt.ylabel('Sale Price')
plt.title("House Price")
plt.show()
```

# Split data into Train and Test Data

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

# Plotting graph on training data

```python
#plt.scatter(X_train, y_train, color = "red")
#plt.plot(X_train, regressor.predict(X_train), color = "green")
```

# Train the algorithm 

```python
regressor=LinearRegression()
regressor.fit(X_train,y_train)
```

# Retrive the Intrecept

```python
print(regressor.intercept_)
```

# Retrive the Slope

```python
print(regressor.coef_)
```

# Predicted Values

```python
y_pred=regressor.predict(X_test)
y_pred
```

```python
plt.scatter(X,y)
plt.plot([min(X),max(X)],[min(y_pred),max(y_pred)],color='red')
plt.show()
```

 # Actual Values

```python
y_test
```

# Evaluating Model

```python
from sklearn import metrics

print("Mean Absoulet Error:-",metrics.mean_absolute_error(y_test,y_pred))
print("Mean Squared Error:-",metrics.mean_squared_error(y_test,y_pred))
print(" Root Mean Square Error:-",np.sqrt(metrics.mean_absolute_error(y_test,y_pred)))
```
